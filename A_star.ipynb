```python
class PriorityQueue:
    def __init__(self):
        self.items = []

    def isEmpty(self):
        return self.items == []

    def enqueue(self, item):
        if self.isEmpty():
            self.items.insert(0, item)
        else:
            index = 0
            for node in self.items:
                if (int(item.path_cost) + int(item.hcost)) < (int(node.path_cost) + int(node.hcost)):
                    index += 1
                    continue
                else:
                    return
            self.items.insert(index, item)

    def dequeue(self):
        return self.items.pop()

    def size(self):
        return len(self.items)

    def contains(self, state):
        for index, node in enumerate(self.items):
            if state == node.state:
                return node
        return False
    
    def replace(self, current_node, new_node):
        for index, node in enumerate(self.items):
            if(current_node.state == node.state) and ((int(current_node.path_cost) + int(current_node.hcost)) > (int(new_node.path_cost) + int(new_node.hcost))):
                self.items[index] = new_node
                break
```

```python
class Node:
    def __init__(self, s, c=0, h=0, p=None):
        self.state = s
        self.path_cost = c
        self.hcost = h
        self.parent = p

    def __repr__(self): 
        return '<{}, {}>'.format(self.state, self.path_cost)
    
    def __lt__(self, other): 
        return self.path_cost < other.path_cost
```

```python
class Problem:
    def __init__(self, s, q, m, h):
        self.initial_state = s
        self.goal_state = q
        self.model = m
        self.hmodel = h

    def goal_test(self, s):
        if self.goal_state == s:
            return True
        else:
            return False

    def actions(self, s):
       return self.model.get(s, False)

    def hcost(self, s):
        return self.hmodel.get(s, False)
```

```python
def A_star(problem):
    node = Node(problem.initial_state)
    
    frontier = PriorityQueue()
    frontier.enqueue(node)
    explored = set()

    while frontier.size() > 0:
        node = frontier.dequeue()
        if problem.goal_test(node.state):
            return node
        explored.add(node.state)
        cities = problem.actions(node.state)
        if cities:
            for (city, cost) in problem.actions(node.state):
                child = Node(city, node.path_cost + cost, problem.hcost(city), node)
                if (child.state not in explored) and (not frontier.contains(child.state)):
                    frontier.enqueue(child)
                elif frontier.contains(child.state):
                    hnode = frontier.contains(child.state)
                    frontier.replace(hnode, child)
    return False
```

```python
prob = Problem('Sibiu','Bucharest', map, hMap)
solution = A_star(prob)
print(solution)
```
