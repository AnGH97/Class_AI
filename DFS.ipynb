```python
class Stack:
    def __init__(self):
        self.items = []
    
    def isEmpty(self):
        return self.items == []

    def __len__(self):
        return len(self.top)
    
    #스택 내부 자료를 string으로 변환하여 반환
    #[::1] 처음부터 끝까지 한칸 간격으로
    def __str__(self):      
        return str(self.top[::1])

    def push(self, item):
        self.items.insert(0, item)

    #들어온 원소의 역순으로 원소를 꺼낸다.
    def pop(self):
        return self.items.pop(-1)

    def clear(self):
        self.items = []
    
    def size(self):
        return len(self.items)

    def contains(self, state):
        for node in self.items:
            if state == node.state:
                return True
        return False
```

```python
class Node:
    def __init__(self, s, p=None, v=0):
        self.state = s
        self.utility = v
        self.parent = p

    def __str__(self):
        return 'State = ' +self.state
```

```python
class Problem:
    def __init__(self, s, q, m):
        self.initial_state = s
        self.goal_state = q
        self.model = m

    def goal_test(self, s):
        if self.goal_state == s:
            return True
        else:
            return False

    def actions(self, s):
       return self.model.get(s, False)
    
```

```python
def dfs(problem):
    node = Node(problem.initial_state)
    if problem.goal_test(node.state):
        return node
    
    frontier = Stack()
    frontier.push(node)
    explored = set()

    while frontier.size() > 0:
        node = frontier.pop()
        explored.add(node.state)
        cities = problem.actions(node.state)
        print(cities)
        if cities:
            for city in problem.actions(node.state):
                child = Node(city, node)
                if (child.state not in explored) and (not frontier.contains(child.state)):
                    if problem.goal_test(child.state) :
                        return child
                    else:
                        frontier.push(child)
    return False
```

```python
prob = Problem('A', 'L', map)
solution = dfs(prob)
print(solution)
```
