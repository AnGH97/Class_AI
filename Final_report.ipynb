```python
def CL(clauses):
    c = []
    list1 = []
    for literal in clauses:
        list1 = list(literal)
        #print(list1)
        #print(len(list1))
        for i in range(1, len(list1)+(len(list1)-1), 2):
            if(len(list1) == 1):
                continue
            else:
                list1.insert(i, 'or')
        #print("list: ",list1)
        c.append(list1)
        #print("*****************************************")
    #print(c)
    #print(len(c))
    for j in range(1, len(c)+len(c), 2):
        c.insert(j, ['and'])
    return c
```
