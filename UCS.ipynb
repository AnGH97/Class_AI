```python
class PriorityQueue:
    def __init__(self):
        self.items = []

    def isEmpty(self):
        return self.items == []

    def enqueue(self, item):
        if self.isEmpty():
            self.items.insert(0, item)
        else:
            index = 0
            for node in self.items:
                if item.path_cost < node.path_cost:
                    index += 1
                    continue
                else:
                    break
            self.items.insert(index, item)

    def dequeue(self):
        return self.items.pop()

    def size(self):
        return len(self.items)

    def contains(self, state):
        for index, node in enumerate(self.items):
            if state == node.state:
                return node
        return False
    
    def replace(self, current_node, new_node):
        for index, node in enumerate(self.items):
            if(current_node.state == node.state) and (current_node.path_cost > new_node.path_cost):
                self.items[index] = new_node
                break
```

```python
class Node:
    def __init__(self, s, c=0, p=None):
        self.state = s
        self.path_cost = c
        self.parent = p

    def __repr__(self): 
        return '<{}, {}>'.format(self.state, self.path_cost)
    
    def __len__(self): return 0 if self.parent is None else (1 + len(self.parent))
    
    def __lt__(self, other): 
        return self.path_cost < other.path_cost
```

```python
class Problem:
    def __init__(self, s, q, m):
        self.initial_state = s
        self.goal_state = q
        self.model = m

    def goal_test(self, s):
        if self.goal_state == s:
            return True
        else:
            return False

    def actions(self, s):
       return self.model.get(s, False)
```

```python
def UCS(problem):
    node = Node(problem.initial_state)
    
    frontier = PriorityQueue()
    frontier.enqueue(node)
    explored = set()

    while frontier.size() > 0:
        node = frontier.dequeue()
        if problem.goal_test(node.state):
            return node
        explored.add(node.state)
        cities = problem.actions(node.state)
        if cities:
            for (city, cost) in problem.actions(node.state):
                child = Node(city, node.path_cost + cost, node)
                if (child.state not in explored) and (not frontier.contains(child.state)):
                    frontier.enqueue(child)
                elif frontier.contains(child.state):
                    hnode = frontier.contains(child.state)
                    frontier.replace(hnode, child)
    return False
```

```python
prob = Problem('Sibiu','Bucharest', map)
solution = UCS(prob)
print(solution)
```
