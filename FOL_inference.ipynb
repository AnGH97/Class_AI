```python
from nltk.inference.resolution import *
from nltk.sem import logic
from nltk.sem.logic import *
logic._counter._value = 0
read_expr = logic.Expression.fromstring
```

```python
p1 = read_expr('all x.(man(x) -> mortal(x))')
p2 = read_expr('man(Socrates)')
c = read_expr('mortal(Socrates)')
logic._counter._value = 0

tp = ResolutionProverCommand(c, [p1,p2])
tp.prove()
print(tp.proof())
```

```python
r1 = read_expr('all x.(King(x) & Greedy(x) -> Evil(x))')
r2 = read_expr('King(John)')
r3 = read_expr('all y.(Greedy(y))')
r4 = read_expr('Brother(Richard, John)')
c = read_expr('Evil(John)')

tp = ResolutionProverCommand(c, [r1,r2, r3, r4])
tp.prove()
print(tp.prove())
print(tp.proof())
```

```python
#1번
p11 = read_expr('Occupation(Emily, Surgeon)')
p12 = read_expr('all p.Occupation(p, Surgeon) -> exists q.Occupation(q, Lawyer) & Customer(p, q)')
c = read_expr('exists x.Customer(Emily, x) and Occupation(x, Lawyer)')
#logic._counter._value = 0

tp = ResolutionProverCommand(c, [p11,p12])
tp.prove()
print(tp.proof())
```

```python
#2번
p21 = read_expr('Occupation(Emily, Surgeon) | Occupation(Emily, Lawyer)')
p22 = read_expr('all p.Occupation(p, Surgeon) -> exists q.Occupation(q, Lawyer) & Customer(p, q)')
c = read_expr('exists x.Customer(Emily, x) and Occupation(x, Lawyer)')
logic._counter._value = 0

tp = ResolutionProverCommand(c, [p21,p22])
tp.prove()
print(tp.proof())
```

```python
#3번
r11 = read_expr('Occupation(Emily, Surgeon)')
r12 = read_expr('all p.Occupation(p, Surgeon) -> exists q.Occupation(q, Lawyer) & Customer(p, q)')
r13 = read_expr('Occupation(Joe, Lawyer)')
c = read_expr('all x.(Customer(Emily, x) and Occupation(x, Lawyer) -> ANSWER(x))')
logic._counter._value = 0

tp = ResolutionProverCommand(None, [r11,r12, r13, c])
sorted(tp.find_answers())
print(tp.prove())
print(tp.proof())
```
