```python
def CL(clauses):
    c = []
    list1 = []
    for literal in clauses:
        list1 = list(literal)
        for i in range(1, len(list1)+(len(list1)-1), 2):
            if(len(list1) == 1):
                continue
            else:
                list1.insert(i, 'or')
        c.append(list1)
    for j in range(1, len(c)+len(c), 2):
        c.insert(j, ['and'])
    return c
```

```python
def plresolveL(pairs):
    literal = []
    new4 = pairs.copy()
    for i in range(0, len(new4)):
        for j in range(0, len(new4[i])):
            for k in range(0, len(new4[i])):
                if(new4[i][j] != new4[i][k]) and (new4[i][j] in new4[i][k]) and ('-' in new4[i][k]):
                    literal.append(new4[i][j])
                    literal.append(new4[i][k])
    return literal
```
