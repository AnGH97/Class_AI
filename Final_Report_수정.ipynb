```python
def CL(clauses):
    c = []
    list1 = []
    for literal in clauses:
        list1 = list(literal)
        for i in range(1, len(list1)+(len(list1)-1), 2):
            if(len(list1) == 1):
                continue
            else:
                list1.insert(i, 'or')
        c.append(list1)
    for j in range(1, len(c)+len(c), 2):
        c.insert(j, ['and'])
    return c
```
