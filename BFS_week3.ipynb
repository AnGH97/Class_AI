```python
class Queue:
    def __init__(self):
        self.items = []

    def isEmpty(self):
        return self.items == []

    def enqueue(self, item):
        self.items.insert(0, item)

    def dequeue(self):
        return self.items.pop()

    def size(self):
        return len(self.items)

    def contains(self, state):
        for node in self.items:
            if state == node.state:
                return True
        return False
```

```python
class Node:
    def __init__(self, s, p=None, c=0):
        self.state = s
        self.path_cost = c
        self.parent = p

    def __str__(self):
        return 'State = ' +self.state
```

```python
class Problem:
    def __init__(self, s, q, m):
        self.initial_state = s
        self.goal_state = q
        self.model = m

    def goal_test(self, s):
        if self.goal_state == s:
            return True
        else:
            return False

    def actions(self, s):
       return self.model.get(s, False)
```

```python
def bfs(problem):
    node = Node(problem.initial_state)
    if problem.goal_test(node.state):
        return node
    
    frontier = Queue()
    frontier.enqueue(node)
    explored = set()

    while frontier.size() > 0:
        node = frontier.dequeue()
        explored.add(node.state)
        cities = problem.actions(node.state)
        print(cities)
        if cities:
            for city in problem.actions(node.state):
                child = Node(city, node)
                if (child.state not in explored) and (not frontier.contains(child.state)):
                    if problem.goal_test(child.state) :
                        return child
                    else:
                        frontier.enqueue(child)
    return False
```

```python
prob = Problem('A', 'L', map)
solution = bfs(prob)
print(solution)
```
