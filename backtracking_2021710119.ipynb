```python
import random
```

```python
map = {
"WA":["NT", "SA"],
"NT": ["WA", "Q", "SA"],
"Q": ["NT", "SA", "NSW"],
"SA": ["WA", "NT", "Q", "NSW", "V"],
"NSW": ["Q", "SA", "V"],
"V": ["SA", "NSW"]
}
```

```python
def color_select():
    color = ["red", "green", "blue"]
    coloring = random.sample(color, 1)
    return coloring
```

```python
def neighbor_dfs(vertex, visited, color, scolor):
    for i in map[vertex]:
        for j in range(len(visited)):
            if (i == visited[j]) and (color == scolor[j]):
                return True
    return False
```

```python
def stack_dfs(start_vertex):
    visited = []
    result = []
    scolor = []
    num = 0
    stack = [(start_vertex)]
    color = ["red", "green", "blue"]
    while stack:
        vertex = stack.pop()
        if vertex not in visited:
            color = color_select()
            answer = (vertex, color)
            if num > 0:
                while(True):
                    if neighbor_dfs(vertex, visited, color, scolor) == True:
                        print(color)
                        print(num)
                        color = color_select()
                        answer = (vertex, color)
                    elif neighbor_dfs(vertex, visited, color, scolor) == False:
                        break
            visited.append(vertex)
            scolor.append(color)
            result.append(answer)
            num += 1
            for item in map[vertex]:
                stack.append(item)
    return result
```
