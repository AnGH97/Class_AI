```python
import random
import copy
```

```python
def gen_individual(length):                                 #10개의 chromosome으로 구성된 1개의 population 생성
    population = []
    for i in range(0, 4):
        genes=[]
        while len(genes)<length:
            sampleSize=min(length-len(genes),len(geneSet))
            genes.extend(random.sample(geneSet,sampleSize))
        population.append(''.join(genes))
    return population

def get_fitness(guess, target):
    fitness = 0
    for expected, actual in zip(target, guess):
        if expected == actual:                      #위치와 문자가 모두 동일한 경우
            fitness += 5    
        elif actual in target:                      #위치는 동일하지 않고 문자만 같은 경우
            fitness += 1
    return fitness
    #return sum(1 for expected,actual in zip(target,guess) if expected==actual)
    #숫자와 위치가 동일하다면 +1, 위치는 같지 않고 문자만 같다면 더하지 않는다.

def mutate(parent):
    index = random.randrange(0,len(parent))
    childGenes = list(parent)
    newGene,alternate = random.sample(geneSet,2)
    childGenes[index] = alternate if newGene == childGenes[index] else newGene
    return''.join(childGenes)

def display(guess, target):
    fitness = get_fitness(guess, target)
    print('{}\t{}'.format(guess,fitness))

```

```python
def _generate_child(population, target):
    child_list = []
    fitness_percent_list = []
    fitness_accum_list = []
    fitness_sum = 0

    for parent in population:
        fitness_sum += get_fitness(parent, target)

    for parent in population:
        fitness_percent_list.append(get_fitness(parent, target) / fitness_sum)

    fitness_sum = 0
    for fitness_percent in fitness_percent_list:
        fitness_sum += fitness_percent
        fitness_accum_list.append(fitness_sum)

    # Selection
    for i in range(0, 4):
        rand = random.random()
        before = 0
        for j in range(0, len(fitness_accum_list)):
            if rand > before and rand <= fitness_accum_list[j]:
                child_list.append(copy.deepcopy(population[j]))
                break
            before = fitness_accum_list[j]
```
